## 04-3 게임 개발

캐릭터가 있는 장소는 1x1 크기의 정사각형으로 이뤄진 NxM 크기의 직사각형.
각각의 칸은 육지 또는 바다이다. 캐릭터는 동서남북 중 한 곳을 바라본다.

맵 각 칸은 (A,B)로 나타낼 수 있고, A는 북쪽으로부터 떨어진 칸의 개수, B는 서쪽으로부터 떨어진 칸의 개수.

캐릭터는 상하좌우로 움직일 수 있고, 바다로 되어 있는 공간에는 갈 수 없다.

캐릭터의 움직임을 설정하기 위해 정해 놓은 메뉴얼

1. 현재 위치에서 현재 방향을 기준으로 왼쪽 방향(반시계 방향으로 90도 회전한 방향)부터 차례대로 갈 곳을 정한다.

2. 캐릭터의 바로 왼쪽 방향에 아직 가보지 않은 칸이 존재한다면, 왼쪽 방향으로 회전한 다음 왼쪽으로 한 칸을 전진한다.
왼쪽 방향에 가보지 않은 칸이 없다면, 왼쪽 방향으로 회전만 수행하고 1단계로 돌아간다.

3. 만약 네 방향 모두 이미 가본 칸이거나 바다로 되어 있는 칸인 경우에는, 
바라보는 방향을 유지한 채로 한 칸 뒤로 가고 1단계로 돌아간다.
단. 이때 뒤쪽 방향이 바다인 칸이라 뒤로 갈 수 없는 경우에는 움직임을 멈춘다.

입력조건
- 첫째 줄에 맵의 세로 크기 N과 가로 크기 M을 공백으로 구분하여 입력 (3 <= N,M <= 50)
- 둘째 줄에 게임 캐릭터가 있는 칸의 좌표 (A,B)와 바라보는 방향 d가 서로 공백으로 구분하여 주어짐
방향 d의 값으로는 다음고 같이 4가지가 존재
    - 0 : 북쪽
    - 1 : 동쪽
    - 2 : 남쪽
    - 3 : 서쪽
- 세째 줄부터 육지인지 바다인지에 대한 정보가 주어진다. N개의 줄에 맵의 상태가 북쪽부터 남쪽 순서대로,
각 줄의 데이터는 서쪽부터 동쪽 순서대로 주어진다. 맵의 외곽은 항상 바다로 되어 있다.
    - 0 : 육지
    - 1 : 바다
- 처음에 게임 캐릭터가 위치한 칸의 상태는 항상 육지

출력조건
- 첫째 줄에 이동을 마친 후 캐릭터가 방문한 칸의 수를 출력

입력 예시
```
4 4      # 4x4 맵 생성
1 1 0    # (1,1) , 북쪽(0) 바라보고 서 있는 캐릭터
1 1 1 1  # 첫 줄은 모두 바다
1 0 0 1  # 둘째 줄은 바다/육지/육지/바다
1 1 0 1  # 셋째 줄은 바다/바다/육지/바다
1 1 1 1  # 넷째 줄은 모두 바다
```
출력 예시
```
>> 3
```

````python
'''
1 1 1 1
1 0(A^) 0 1
1 1 0 1
1 1 1 1

-------

북 (-1,0)
동 (0,1)
남 (1,0)
서 (0,-1)

현재 방향 : 0 북 1 동 2 남 3 서 -1 서
가본 칸 mark -> +1
회전 turn
캐릭터가 가본 칸의 수 cnt

움직임 메뉴얼
1) 왼쪽으로 회전 -> 안 가본 경우 -> 그 방향으로 직진, turn = 0
2) 왼쪽으로 회전 -> 가본 경우 -> 왼쪽으로 회전, turn += 1

turn = 4
3) 모든 방향으로 다 가본 경우, 바다로 된 칸 -> 뒤로 갈 수 있다면 뒤로 감
4) 모든 방향으로 다 가본 경우, 바다로 된 칸 -> 뒤로 갈 수 없다면 break

--------

(1,1)
1. 왼쪽으로 회전 -> Map = 1 -> turn = 1
2. 왼쪽으로 회전 -> Map = 1 -> turn = 2
3. 왼쪽으로 회전 -> Map = 0 , mark = 0 -> (1,2) 동쪽, turn = 0
4. 왼쪽으로 회전 -> Map = 1 -> turn = 1
5. 왼쪽으로 회전 -> Map = 0, mark = 1 -> turn = 2
6. 왼쪽으로 회전 -> Map = 0, mark = 0 -> (2,2) 북쪽, turn = 0
7. 모든 방향으로 다 가본 경우 -> 뒤로 가능 - (1,2), turn = 0
8. 모든 방향으로 다 가본 경우 -> 뒤로 불가 -> break

'''
# 현재 위치에서 현재 방향을 기준으로 왼쪽 방향(반시계 방향으로 90도 회전한 방향)부터 차례때로 갈 곳을 정한다.

# 캐릭터의 바로 왼쪽 방향에 아직 가보지 않은 칸이 존재한다면, 
# 왼쪽 방향으로 회전한 다음 왼쪽으로 한 칸을 전진한다. 
# 왼쪽 방향에 가보지 않은 칸이 없다면, 왼쪽 방향으로 회전만 수행하고 1단계로 돌아간다.

# 만약 네 방향 모두 이미 가본 칸이거나 바다로 되어 있는 칸인 경우에는, 
# 바라보는 방향을 유지한 채로 한 칸 뒤로 가고 1단계로 돌아간다. 
# 단. 이때 뒤쪽 방향이 바다인 칸이라 뒤로 갈 수 없는 경우에는 움직임을 멈춘다

# 캐릭터가 방문한 칸의 수
````

````python
N, M = map(int, input().split())  # 맵 크기
A, B, d = map(int, input().split())  # 현재 좌표 (A,B), d : 방향
Map = []
for _ in range(N):
    Map.append(list(map(int, input().split())))   # 1.

# print(Map)
# [[1, 1, 1, 1], [1, 0, 0, 1], [1, 1, 0, 1], [1, 1, 1, 1]]

# 북 동 남 서  0 1 2 3
route = [(-1,0),(0,1),(1,0),(0,-1)]  # 2.

# 가본 칸
mark = [[0 for _ in range(M)] for _ in range(N)]  # 3.
# print(mark)

mark[A][B] = 1 # 현재 위치 => 가본 경우 이므로 1  # 4.
# print(mark)

turn = 0  # 회전 횟수
cnt = 0 # 가본 횟수

while True:
    # 왼쪽으로 회전   # 5.
    d -= 1
    if d == -1:  # 북 -> 서로 돌 때  -1이 되면 서쪽 방향인 3
        d = 3
    
    # 안 가본 경우 -> 그 방향으로 직진 / turn 초기화  # 6.
    if Map[A+route[d][0]][B+route[d][1]] == 0 and mark[A+route[d][0]][B+route[d][1]] == 0:
        mark[A+route[d][0]][B+route[d][1]] = 1
        A += route[d][0]
        B += route[d][1]
        cnt += 1
        turn = 0
    
    # 가본 경우 -> 회전만 / 직진 X / turn += 1   # 7.
    else:
        turn += 1
        
    # 모든 방향으로 다 가본 경우   # 8.
    if turn == 4:
        # 뒤로 갈 수 있다면 뒤로 감  # 9.
        # 남쪽방향 -> 뒤로 -> -2
        if Map[A+route[d-2][0]][B+route[d-2][1]] == 0:
            A += route[d-2][0]
            B += route[d-2][1]
            turn = 0
            cnt += 1
        # 뒤로 갈 수 없다면 break
        else:
            break
#     print(mark)

print(cnt)
````

### 풀이
1. 전체 맵을 나타내기 위해 리스트 받아서 배열 형태로 나타내기
2. 갈 수 있는 방향의 경우의 수 나타내기
3. 가본 칸을 체크하기 위해 전체 맵의 크기만큼 0으로 초기화된 배열
4. 현재 위치는 이미 가본 칸이므로 mark[A][B]은 1임
5. 현재 방향에서 왼쪽으로 방향을 돌리면 d는 1씩 뺀 상태 

    북, 동, 남 ,서는 순서대로 0, 1, 2, 3임

    서쪽에서 왼쪽으로 돌면 남쪽이고 3 -1 = 2

    남쪽에서 왼쪽으로 돌면 동쪽이고 2 -1 = 1

    북쪽에서 서쪽으로 돌리면 0에서 -1이 되고 서쪽인 3으로 고정
6. 움직임 메뉴얼 1번. 왼쪽으로 회전 -> 안 가본 경우 -> 그 방향으로 직진, turn = 0

    왼쪽으로 회전한 후 (d -=1) 안 가본 경우, 그쪽 방향으로 갈 수 있는 경우라면
    
    mark가 0이면 안 가본 경우이고, Map이 0이면 육지이므로 갈 수 있으므로
    
    그쪽 방향으로 갔단 의미로 mark를 1을 추가하고
    
    A와 B를 route[d]로 업데이트
    
    ```
    3. 왼쪽으로 회전 -> Map = 0 , mark = 0 -> (1,2) 동쪽, turn = 0
    6. 왼쪽으로 회전 -> Map = 0, mark = 0 -> (2,2) 북쪽, turn = 0
    ```
    (1,1) -> + (동쪽 = (0,1)) = (1,2) 
    
    turn을 0으로 초기화
    
    한칸 이동했으므로 cnt 1추가
    
7. 6의 경우가 아니라면 turn만 1추가
    ```
    1. 왼쪽으로 회전 -> Map = 1 -> turn = 1
    2. 왼쪽으로 회전 -> Map = 1 -> turn = 2
    4. 왼쪽으로 회전 -> Map = 1 -> turn = 1
    5. 왼쪽으로 회전 -> Map = 0, mark = 1 -> turn = 2
    ```

8. 모든 경우를 다 가본 경우라면 turn이 4인 상태

9. 뒤로 갈 수 있다면 뒤로 이동. d가 -2 더해진다면 반대 방향으로 이동

    이 경우도 한 칸 이동한 것이므로 cnt 1 추가
    
    한 칸 이동했으므로 turn도 0으로 초기화
    ```
    7. 모든 방향으로 다 가본 경우 -> 뒤로 가능 - (1,2), turn = 0
    ```

### 책 답안 예시

````python
# N, M을 공백으로 구분하여 입력받기
n, m = map(int, input().split())

# 방문한 위치를 저장하기 위한 맵을 생성하며 0으로 초기화
d = [[0] * m for _ in range(n)]

# 현재 캐릭터의 X 좌표, Y 좌표, 방향을 입력받기
x, y, direction = map(int, input().split())
d[x][y] = 1 # 현재 좌표 방문 처리

# 전체 맵 정보를 입력받기
array = []
for i in range(n):
    array.append(list(map(int, input().split())))
    
# 북, 동, 남 ,서 방향 정의
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 왼쪽으로 회전
def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3
    
# 시물레이션 시작
count = 1
turn_time = 0

while True:
    # 왼쪽으로 회전
    turn_left()
    nx = x + dx[direction]
    ny = y + dy[direction]
    
    # 회전한 이후 정면에 가보지 않은 칸이 존재하는 경우 이동
    if d[nx][ny] == 0 and array[nx][ny] == 0:
        d[nx][ny] = 1
        x = nx
        y = ny
        count += 1
        turn_time = 0
        continue
    
    # 횢너한 이후 정면에 가보지 않은 칸이 없거나 바다인 경우
    else:
        turn_time += 1
    
    # 네 방향 모두 갈 수 없는 경우
    if turn_time == 4:
        nx = x - dx[direction]
        ny = y - dy[direction]
        
        # 뒤로 갈 수 있다면 이동하기
        if array[nx][ny] == 0:
            x = nx
            y = ny
        # 뒤가 바다로 막혀있는 경우
        else:
            break
        turn_time = 0

#정답 출력
print(count)
````
